```python
from pathlib import Path
import re

src = Path("/content/basic_english_850.txt")
text = src.read_text(encoding="utf-8")

lines = text.splitlines()

words = []

for line in lines:
    line_strip = line.strip()
    if not line_strip:
        continue
    if "http" in line_strip:
        continue
    if "OGDEN" in line_strip or "BASIC ENGLISH" in line_strip:
        continue
    if line_strip.startswith("Word List"):
        continue
    if line_strip.startswith("OPERATIONS"):
        continue
    if line_strip.startswith("THINGS"):
        continue
    if line_strip.startswith("QUALITIES"):
        continue
    if line_strip.isupper():
        continue

    for token in line_strip.split():
        token = token.strip(" ,.;:-–—()")
        if not token:
            continue
        parts = re.split(r"/", token)
        for p in parts:
            p = p.strip().lower()
            if not p:
                continue
            if not re.fullmatch(r"[a-z]+", p):
                continue
            words.append(p)



# dedupe and sort
unique_words = sorted(set(words))
print("Total unique items:", len(unique_words))
unique_words[:30]
```

```python
out = Path("/content/basic_english_850.txt")
out.write_text("\n".join(unique_words), encoding="utf-8")
print("Wrote:", out)
```
